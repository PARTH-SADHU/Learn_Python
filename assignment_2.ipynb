```python
import pandas as pd
import numpy as np
```

```python
X=pd.read_csv("document.csv", names=["name","income"],skiprows=[0])
X
```

```python
# Describe function gives the description of data,for numerical data it gives count,mean and standard deviation with minimum and maximum values.
X.describe()
```

```python
# Using income keyword we can access that particular named column and by using quantile method we can get 10% or 25% values:
forty_fifth=X.income.quantile(0.45)
print(forty_fifth)
point_ten=X.income.quantile(0.10)
print(point_ten)
```

```python
max_income=X.income.quantile(0.99)
print(max_income)
```

```python
#Removing max value usinng percentile:
X_No_maxval=X[X.income<=max_income]
print(X_No_maxval)
```
